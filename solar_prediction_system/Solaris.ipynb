```python
"""
태양광 발전량 예측 시스템 메인 실행 파일
"""
import os
from config import get_config
from web.app import create_app
from utils.file_utils import ensure_directories

def setup_environment():
    """환경 설정 및 초기화"""
    config = get_config()
    
    # 필요한 디렉토리 생성
    ensure_directories()
    
    # ngrok 설정 (개발 환경에서만)
    if config.FLASK_ENV == 'development' and config.NGROK_AUTH_TOKEN:
        try:
            from pyngrok import ngrok
            ngrok.set_auth_token(config.NGROK_AUTH_TOKEN)
            public_url = ngrok.connect(config.PORT, region="ap")
            
            print(f"\n🌍 Public URL: {public_url}")
            print(f"📊 태양광 발전량 예측: {public_url}")
            print(f"🔥 히트맵 보기: {public_url}/heatmap")
            print(f"⚡ 시스템 설계: {public_url}/system_designer")
            print(f"📥 데이터 다운로드: {public_url}/download")
            
        except Exception as e:
            print(f"\n⚠️ ngrok 연결 실패: {str(e)}")
            print(f"🌍 로컬에서 접속하세요: http://127.0.0.1:{config.PORT}")
    else:
        print(f"\n🌍 로컬에서 접속하세요: http://127.0.0.1:{config.PORT}")
```

```python
"""메인 함수"""
# 환경 설정
setup_environment()

# Flask 앱 생성
app = create_app()
config = get_config()

# 서버 실행
app.run(
    host='0.0.0.0',
    port=config.PORT,
    debug=config.FLASK_DEBUG
)
```
